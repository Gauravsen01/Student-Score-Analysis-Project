### Gender Distribution - This in not the bar graph but it's a count plot which comes under seaborn library. 
### countplot() method is used to Show the counts of observations in each categorical bin using bars.

```python
plt.figure(figsize =(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

### From the above chart we have analyzed that the number of female in the data is more than the number of male. 

### Now to analyze how parents education impacting to the students score.
### We will use groupby function to check the same.

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print (gb)
```

### Here we can see all the mean values of every subject along with the parents education.

### By using heat map we can understand the distribution of data in a better manner.
### To plot the heat map, We have to use Seaborn library as sns.heatmap(data).
### Also to show values of each colour we have used annot = TRUE.

```python
sns.heatmap(gb, annot = True)
plt.title("ralationship between parents education and student score")
plt.show()
```

### From the above chart we have concluded that the education of the parents have a good impact on student score. 

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print (gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title('Relationship between parents marital status and student score')
plt.show()
```

### From the above chart we have conclude that there is no/negligible impact on the student score due to their parents marital status.

### Distribution of Ethnic Groups
### We will calculate the percentage distribution of each ethnic group with the help of Pie chart and For that first we need to find out the count of each group.

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

print(groupA, groupB, groupC, groupD, groupE)
```

### Have created a pie chart of Ethnic group and the values of each group alloted to the variable name mylist.

### created new variable name as l, To put lables in pie chart we have added labels = l in plt.pie followed by autopct = "%1.2f%%" to show the percentage in the pie chart. 

### we call this percentage sign (%) as Modulus.

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ("groupA", "groupB", "groupC", "groupD", "groupE")

mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.title("Distribution of Ethnic Group")
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.show
```
